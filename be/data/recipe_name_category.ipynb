```python
import pandas as pd
from openai import OpenAI
import time

client = OpenAI(api_key="sktax-XyeKFrq67ZjS4EpsDlrHHXV8it", base_url="https://guest-api.sktax.chat/v1")

# CSV 읽기
recipes = pd.read_csv("recipes_has_url.csv")

def clean_title_with_ai(title: str) -> str:
    prompt = (
        f"다음 요리 제목에서 불필요한 설명, 괄호 속 내용, 수식어를 제거하고 "
        f"순수한 음식 이름만 남겨줘.\n"
        f"입력: {title}\n"
        f"출력:"
    )

    completion = client.chat.completions.create(
        model="ax4",
        messages=[
            {"role": "user", "content": prompt}
        ]
    )

    return completion.choices[0].message.content.strip()

# 한 행씩 처리
cleaned_titles = []
for idx, row in recipes[:30].iterrows():
    cleaned = clean_title_with_ai(row["title"])
    cleaned_titles.append(cleaned)
    
    print(f"{idx+1}/{len(recipes)} 처리 완료: {cleaned}")

```
