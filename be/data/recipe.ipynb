## 카테고리 추가

```python
import pandas as pd
from openai import OpenAI
import time

client = OpenAI(api_key="sktax-XyeKFrq67ZjS4EpsDlrHHXV8it", base_url="https://guest-api.sktax.chat/v1")
```

```python
# AI에게 카테고리 분류 요청
def categorize_with_ai(title):
    """
    title을 입력하면 AI가 한식, 중식, 양식, 디저트, 음료 등으로 분류
    """
    prompt = f"다음 음식 제목을 보고 [한식, 중식, 양식, 일식, 간식, 기타] 중 가장 적절한 카테고리 하나만 선택하세요.\n선택한 카테고리의 단어만 출력하고, 다른 말은 절대 하지 마세요.\n\n음식 제목: {title}"
    completion = client.chat.completions.create(
        model="ax4",
        messages=[{"role": "user", "content": prompt}]
    )
    # AI 응답에서 텍스트 추출
    category = completion.choices[0].message.content.strip()
    return category
```

## 시간, 인분 int화

```python
import pandas as pd
```

```python
recipes = pd.read_csv('recipes_with_categories.csv')
```

```python
# 널 제거
len(recipes.dropna(subset=["title", "servings", "time", "difficulty", "materials", "tools", "steps", "danger", "category"]))
# null인 행 개수
# len(recipes[recipes["difficulty"].isnull()])
```

```python
recipes.dropna(subset=["title", "servings", "time", "difficulty", "materials", "tools", "steps", "danger", "category"], inplace=True)
```

```python
# 정규표현식 패턴
pattern = r"^(?P<ingredients>[가-힣a-zA-Z\s]+?)\s*(?:(?P<quantity>\d+(?:/\d+)?)(?P<unit>[가-힣a-zA-Z]+|적당량|약간))?$"

recipes_ingredients = (
    recipes
    .assign(ingredients=lambda df: df['materials'].str.split(','))  # 재료 리스트 분리
    .explode('ingredients')                                         # 행으로 펼치기
    .assign(ingredients=lambda df: df['ingredients'].str.strip())     # 공백 제거
    .reset_index(drop=True)[['title', 'ingredients']]
)

recipes_ingredients = pd.concat([recipes_ingredients['title'], recipes_ingredients['ingredients'].str.extract(pattern)], axis=1)

recipes_ingredients
```

```python
# AI에게 카테고리 분류 요청
def categorize_with_ai(steps):
    """
    steps을 입력하면 AI가 필요한 조리 도구를 분류
    """
    prompt = f"다음 조리 단계를 보고 ['계량도구', '칼', '뒤집개', '필러', '냄비', '국자', '가위', '도마', '전자레인지', '체/거름망', '집게', '주걱', '에어프라이어', '그릇', '프라이팬'] 중 필요한 모든 조리 도구를 선택하세요. \n선택한 조리 도구들을 쉼표(,)로 구분하여 출력하고, 다른 말은 절대 하지 마세요.\n(예: 칼, 가위, 냄비)\n\n조리 단계: {steps}"
    completion = client.chat.completions.create(
        model="ax4",
        messages=[{"role": "user", "content": prompt}]
    )
    # AI 응답에서 텍스트 추출
    tool = completion.choices[0].message.content.strip()
    return tool
```
