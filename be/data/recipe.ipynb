## 카테고리 추가

```python
import pandas as pd
from openai import OpenAI
import time

client = OpenAI(api_key="sktax-XyeKFrq67ZjS4EpsDlrHHXV8it", base_url="https://guest-api.sktax.chat/v1")
```

```python
# AI에게 카테고리 분류 요청
def categorize_with_ai(title):
    """
    title을 입력하면 AI가 한식, 중식, 양식, 디저트, 음료 등으로 분류
    """
    prompt = f"다음 음식 제목을 보고 [한식, 중식, 양식, 일식, 간식, 기타] 중 가장 적절한 카테고리 하나만 선택하세요.\n선택한 카테고리의 단어만 출력하고, 다른 말은 절대 하지 마세요.\n\n음식 제목: {title}"
    completion = client.chat.completions.create(
        model="ax4",
        messages=[{"role": "user", "content": prompt}]
    )
    # AI 응답에서 텍스트 추출
    category = completion.choices[0].message.content.strip()
    return category
```

```python
# CSV 불러오기
recipes = pd.read_csv("recipes_cleaned.csv")

# category 컬럼 초기화
categories = []
```

## 시간, 인분 int화

```python
import pandas as pd
```

```python
recipes = pd.read_csv('recipes_with_categories.csv')
```

```python
# 널 제거
len(recipes.dropna(subset=["title", "servings", "time", "difficulty", "materials", "tools", "steps", "danger", "category"]))
# null인 행 개수
# len(recipes[recipes["difficulty"].isnull()])
```

```python
recipes.dropna(subset=["title", "servings", "time", "difficulty", "materials", "tools", "steps", "danger", "category"], inplace=True)
```

```python
# 허용 값 리스트
allowed_values = ['초급', '아무나', '중급', '고급']
# 조건 필터
recipes = recipes[recipes['difficulty'].isin(allowed_values)]


# '인분'으로 끝나는 행만 유지
recipes = recipes[recipes['servings'].str.endswith('인분', na=False)]
# '인분' 글자 제거
recipes['servings'] = recipes['servings'].str.replace('인분', '', regex=False)


# '분 이내'으로 끝나는 행만 유지
recipes = recipes[recipes['time'].str.endswith('분 이내', na=False)]
# '분 이내' 글자 제거
recipes['time'] = recipes['time'].str.replace('분 이내', '', regex=False)

recipes.head(3)
```

```python
recipes.to_csv("recipes.csv", index=False)
```

## 도구 있는 레시피만 필터링

```python
import pandas as pd
```

```python
recipes = pd.read_csv('recipes.csv')
recipes = recipes.dropna(subset=['title', 'tools', 'materials'])
recipes
```

```python
recipes.shape
```

```python
recipes_tools = (
    recipes.assign(tools=lambda df: df['tools'].str.split(','))
          .explode('tools')
          .assign(tools=lambda df: df['tools'].str.strip())
          .reset_index(drop=True)[['title', 'tools']]
)
recipes_tools[['title', 'tools']]
```

```python
recipes_tools['tools'].unique()
```

```python
recipes_tools.to_csv('recipes_tools.csv', index=False)
```

```python
recipes.head(1)
```

```python
# 정규표현식 패턴
pattern = r"^(?P<ingredients>[^\d]+?)\s*(?P<quantity>\d+(?:/\d+)?)?\s*(?P<unit>[가-힣a-zA-Z]+)?$"

recipes_ingredients = (
    recipes
    .assign(ingredients=lambda df: df['materials'].str.split(','))  # 재료 리스트 분리
    .explode('ingredients')                                         # 행으로 펼치기
    .assign(ingredients=lambda df: df['ingredients'].str.strip())     # 공백 제거
    .reset_index(drop=True)[['title', 'ingredients']]
)

recipes_ingredients = pd.concat([recipes_ingredients['title'], recipes_ingredients['ingredients'].str.extract(pattern)], axis=1)

recipes_ingredients
```

```python
recipes_ingredients.to_csv('recipes_ingredients.csv', index=False)
```
